# Normalisation

Pourquoi normaliser ?
C'est une méthode qui permet de résoudre le problème d'ordre de grandeur des données en ajustant les valeurs des variables pour qu'elles se situent dans un intervalles spécifié, souvent entre 0 et 1, ou autour de la moyenne avec un ecart-type donné.

**Les avantages ? **


*   Comparabilité : permet de comparer avec d'autres modèles de données similaires
*   Optimisation des modélisations en ML


*   Stabilisation des modèles
*   Amélioration de l'interopérabilité

**Les méthodes: **



*   Min-max scalin
*   Standardisation = ramener les données à une valeur standard définie


*   Transformation en percentile

```python
clients = df
```

#Nomalisation Min-max

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Là on demande au code de localiser la colonne qui correspond à l'année de naissance et de voir toutes
#toutes les autres colonnes differentes de celle-ci
xnum = clients.loc[:, clients.columns == "annee_naissance"]
xcat = clients.loc[:, clients.columns != "annee_naissance"]
```

```python
#Initialiser l'échelle min et max
sc = MinMaxScaler()
xnum = pd.DataFrame(sc.fit_transform(xnum), columns=xnum.columns)
xnum

#Résultat: les valeurs sont normalisées entre 0 et 1
```

```python
clients = pd.merge(xnum, xcat, left_index=True, right_index=True)
clients.head()
```
