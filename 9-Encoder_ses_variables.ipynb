Label encoding

```python
#S'assurer d'abnord que le jeux de données ne contient aucune valeur manquante en supprimant toutes celles qui peuvent exister.
client = df.dropna()
client.dtypes
```

```python
client.head()
```

```python
#Convertir les valeurs textuelles en valeurs numériques (f et h)
from sklearn.preprocessing import LabelEncoder

#Création de l'instance
label_encoder = LabelEncoder()
#Appliquer le label encoder pour convertir les valeurs 'f' et 'h' en 0 et 1
client['genre'] = label_encoder.fit_transform(client['genre'])

#Affichier le dataFrame pour vérifier la conversion
client
```

```python
client['genre'].value_counts()
```

One Hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore')

#Appliquer l'encodage sur la colonne 'genre'
encoder_df = pd.DataFrame(encoder.fit_transform(client[['genre']]).toarray())
encoder_df

#Créer un nouveau DataFrame avec les données encodées
```
